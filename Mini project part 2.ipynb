# Elbow Method

```python
#1. Re-scale data
scaler = StandardScaler()
df2_scaled = scaler.fit_transform(df2)
```

```python
df2_scaled.shape
```

```python
first_score = []
for i in range(1,20):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df2_scaled)
    first_score.append(kmeans.inertia_) 
plt.plot (first_score, 'bx')
```

```python
kmeans = KMeans(7)
kmeans.fit(df2_scaled)
labels = kmeans.labels
```
